```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "name":["홍길동","이순신"],
    "age":[20,30]
}
)

df
```

```python
data = df.to_json(orient="records")
print(type(data))


# to_json의 결과는 문자열이다
```

```python
data = df.to_json(orient="columns")
print(type(data))
data
```

```python
data = df.to_json(orient="split")
print(type(data))
data
```
